```python
def hitung_sudut(k1, k2, k3):
    """
    Menghitung sudut di titik k2 yang dibentuk oleh (k1 -> k2) dan (k2 -> k3).
    k1, k2, k3 adalah tuple (x, y) atau array [x, y].
    Rumus: sudut = arccos( (v1 dot v2) / (|v1| * |v2|) )
    Hasil dalam derajat (degree).
    """
    # Membentuk vektor v1 = k1->k2 dan v2 = k2->k3
    # Namun untuk sudut di K2, biasanya v1 adalah (k1 - k2) (arahnya masuk ke k2)
    # dan v2 adalah (k3 - k2) (arahnya keluar dari k2).
    # Keduanya valid, hanya perlu konsisten penggunaan vektor.
    
    v1 = np.array(k1) - np.array(k2)
    v2 = np.array(k3) - np.array(k2)
    
    # Dot product
    dot = np.dot(v1, v2)
    # Magnitude (panjang) masing-masing vektor
    mag_v1 = np.linalg.norm(v1)
    mag_v2 = np.linalg.norm(v2)
    
    # Pastikan tidak membagi dengan nol
    if mag_v1 == 0 or mag_v2 == 0:
        return np.nan  # atau 0, atau definisi lain sesuai kebutuhan
    
    # Sudut dalam radian
    cos_theta = dot / (mag_v1 * mag_v2)
    # Untuk menghindari error floating jika cos_theta > 1 atau < -1 akibat pembulatan
    cos_theta = np.clip(cos_theta, -1.0, 1.0)
    theta_radian = np.arccos(cos_theta)
    
    # Konversi ke derajat
    theta_degree = np.degrees(theta_radian)
    return theta_degree
```

# 5. Perhitungan sudut relatif ke KP1 -> KP2 -> KP3

```python
import numpy as np
def hitung_sudut_horizontal(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    rad = np.arctan2(dy, dx)
    deg = np.degrees(rad)
    return deg if deg >= 0 else deg + 360  # Konversi ke [0°, 360°]

def hitung_sudut_relatif(x1, y1, x2, y2, x3, y3):
    # Vektor pertama: KP1 -> KP2
    v1_x, v1_y = x2 - x1, y2 - y1
    # Vektor kedua: KP2 -> KP3
    v2_x, v2_y = x3 - x2, y3 - y2

    # Dot product
    dot_product = v1_x * v2_x + v1_y * v2_y
    # Magnitude vektor
    magnitude_v1 = np.sqrt(v1_x**2 + v1_y**2)
    magnitude_v2 = np.sqrt(v2_x**2 + v2_y**2)

    # Cegah pembagian oleh nol
    if magnitude_v1 == 0 or magnitude_v2 == 0:
        return 0  # Jika salah satu vektor nol, sudut dianggap 0°

    # Hitung sudut dalam radian dan konversi ke derajat
    rad = np.arccos(dot_product / (magnitude_v1 * magnitude_v2))
    deg = np.degrees(rad)
    
    return deg  # Sudut dalam derajat


# Sudut1: KP1 → KP2 terhadap horizontal
keypoints['angle1_deg'] = keypoints.apply(lambda row: 
    hitung_sudut_horizontal(
        row['pixelx_kp1'], row['pixely_kp1'],  # Titik awal
        row['pixelx_kp2'], row['pixely_kp2']   # Titik akhir
    ),
    axis=1
)

# Sudut2: KP1 → KP2 → KP3 
keypoints['angle2_deg'] = keypoints.apply(lambda row: 
    hitung_sudut_relatif(
        row['pixelx_kp1'], row['pixely_kp1'],  # KP1
        row['pixelx_kp2'], row['pixely_kp2'],  # KP2
        row['pixelx_kp3'], row['pixely_kp3']   # KP3
    ),
    axis=1
)

# Lihat hasil
keypoints
```
